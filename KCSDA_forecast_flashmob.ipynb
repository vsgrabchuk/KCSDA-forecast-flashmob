```python
# Подневные данные
start_date = "toDate('2024-03-15')"
end_date = "toDate('2024-03-21')"

query = f'''
WITH 
    users_activity_t AS (
        SELECT
            toDate(time) AS date,
            user_id,
            countIf(action = 'like') AS user_likes,
            countIf(action = 'view') AS user_views
        FROM {db}.feed_actions
        GROUP BY date, user_id
    ),
    date_stats_t AS (
        SELECT
            date,
            COUNT(DISTINCT user_id) AS uniq_users,
            SUM(user_likes) AS likes,
            SUM(user_views) AS views,
            likes/views AS ctr,
            AVG(user_likes) AS likes_mean,
            AVG(user_views) AS views_mean
        FROM users_activity_t
        WHERE date<toDate(now())
        GROUP BY date
        ORDER BY date
    ),
    date_posts_t AS (
        SELECT
            toDate(time) AS date,
            COUNT(DISTINCT post_id) AS uniq_posts
        FROM {db}.feed_actions
        GROUP BY date
    )

SELECT *
FROM date_stats_t
    LEFT JOIN date_posts_t
        USING date
'''

df = ph.read_clickhouse(query=query, connection=connection)
df.head()
```

*retention*

Требования к данным:
- минимум 3 недели в тренировочной выборке
- чтобы не перегружать информацией, можно посмотреть дни с наибольшим числом зарегистрировавшихся пользователей

```python
# Статистика уникальных пользователей по дням недели за всё время

query = f'''
WITH user_activity AS (
    SELECT DISTINCT
        toDate(time) AS date,
        toDayOfWeek(date) AS dow,
        user_id,
        MIN(date) OVER () AS min_date  -- Минимальная дата в таблице
    FROM {db}.feed_actions
    WHERE date < toDate(now())
),
min_date_t AS (
    -- Минимальная дата, чтобы в таблице было целое число недель
    SELECT toDate(now()) - MAX(DISTINCT floor((date - min_date) / 7) * 7)  AS date
    FROM user_activity
),
dow_uniqs AS (
    SELECT
        dow,
        COUNT(DISTINCT user_id) AS uniq_users
    FROM user_activity
    WHERE date >= (SELECT * FROM min_date_t)
    GROUP BY dow
)

SELECT *
FROM dow_uniqs
'''
sns.barplot(ph.read_clickhouse(query=query, connection=connection), x='dow', y='uniq_users');
plt.grid()
# 1 - пн
```

```python
# Статистика для недели, которя была за 21 день до флешмоба

target_date = "toDate('2024-03-15')"


query = f'''
WITH user_activity AS (
    SELECT DISTINCT
        toDate(time) AS date,
        toDayOfWeek(date) AS dow,
        user_id,
        MIN(date) OVER () AS min_date  -- Минимальная дата в таблице
    FROM {db}.feed_actions
    WHERE date < toDate(now())
),
min_date_t AS (
    -- Минимальная дата, чтобы в таблице было целое число недель
    SELECT toDate(now()) - MAX(DISTINCT floor((date - min_date) / 7) * 7)  AS date
    FROM user_activity
),
dote_uniqs AS (
    SELECT
        date,
        dow,
        COUNT(DISTINCT user_id) AS uniq_users
    FROM user_activity
    WHERE 
        date >= (SELECT * FROM min_date_t)
        AND {target_date} - date >= 21
        AND {target_date} - date < 28
    GROUP BY date, dow
    ORDER BY uniq_users
)

SELECT *
FROM dote_uniqs
'''
target_dates_df = ph.read_clickhouse(query=query, connection=connection)
plt.figure(1)
sns.barplot(target_dates_df, x='date', y='uniq_users');
plt.grid();
plt.xticks(rotation=45);
plt.figure(2)
sns.barplot(target_dates_df, x='dow', y='uniq_users', order=target_dates_df['dow']);
plt.grid();
# 1 - пн
```

```python
# retention за дни с максимальным числом пользователей

target_date = "toDate('2024-03-15')"
date_1 = "toDate('2024-02-20')"
date_2 = "toDate('2024-02-21')"

query = f'''
WITH user_activity AS (
    SELECT DISTINCT
        toDate(time) AS date,
        user_id,
        MIN(date) OVER (PARTITION BY user_id) AS first_date, -- Когда пользователь зашёл в первый раз
        MIN(date) OVER () AS min_date  -- Минимальная дата в таблице
    FROM {db}.feed_actions
    WHERE date < toDate(now())
),
min_date_t AS (
    -- Минимальная дата, чтобы в таблице было целое число недель
    SELECT toDate(now()) - MAX(DISTINCT floor((date - min_date) / 7) * 7)  AS date
    FROM user_activity
),
retention_t_1 AS (
    SELECT
        date,
        first_date,
        COUNT(DISTINCT user_id) AS uniq_users
    FROM user_activity
    WHERE 
        date >= (SELECT * FROM min_date_t)
        AND first_date IN ({date_1}, {date_2})
    GROUP BY date, first_date
),
retention_t_2 AS (
    SELECT
        date,
        first_date,
        uniq_users,
        MAX(uniq_users) OVER (
            PARTITION BY first_date
        ) AS day_0_uniq_users
    FROM retention_t_1
),
retention_t AS (
    SELECT
        date,
        first_date,
        uniq_users / day_0_uniq_users AS retention
    FROM retention_t_2
    ORDER BY first_date, date
)

SELECT *
FROM retention_t
'''
retention_df = ph.read_clickhouse(query=query, connection=connection)
retention_df
```
